```python
cat_count=data['Genre'].value_counts()
cat_count
```

**Ploting Bar Char using matplotlib**
ploting any chart using matplotlib-
1.needs to code more line compare to seaborn
2.Bar chart mainly using for counting numbers
in this below bar graph is univariant ploting like categorical data (Genre)


```python
from matplotlib.lines import lineStyles
x_val=cat_count.index
y_val=cat_count.values
plt.bar(x_val,y_val,color='orange')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Count of Genre')
plt.grid(False)
plt.ylim(0,3500)
plt.show()
```

Count plot using seaborn(same as bar graph)

```python
sns.countplot(data=data,x='Genre',order=data['Genre'].value_counts().index,color='orange',palette='rainbow')
plt.title('Count of Genre')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=80,fontsize=12)
plt.show()
```

**PIE chart**- Pie chart mainly used for analysis of proportion.
its not present in seaborn

```python
sales_data=data[['NA_Sales','EU_Sales','JP_Sales','Other_Sales']]
sales_data.head()
```

```python
region_sales=sales_data.T.sum(axis='columns')
region_sales
```

```python
plt.pie(region_sales,labels=region_sales.index,startangle=90,explode=(0.1,0,0,0),autopct='%1.2f%%')
plt.title('Region Sales')
plt.show()
```
