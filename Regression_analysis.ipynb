Linear Regression Algorithm(simple linear)

simple linear regression is a type of regression algo that models the relationship between a dependent variable and a single independent variable.
y=mx+c
x=slope of the line or coeficient of the x
c=intersept

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import  files
files.upload_file('placement.csv')
```

```python
data=pd.read_csv('placement.csv')
data.head(3)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(8,2))
sns.scatterplot(x='cgpa',y='package',data=data)
plt.show()
```

```python
x=data[['cgpa']]
y=data['package']
x.ndim#for ml modle input must be multidimensional
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
le=LinearRegression()
le.fit(x_train,y_train)

```

```python
le.predict(x_test)
# le.predict([[6.89]])
```

```python
le.score(x_test,y_test)*100
```

```python
# y=mx+c
le.coef_
```

```python
le.intercept_
```

```python
plt.figure(figsize=(8,3))
sns.scatterplot(x='cgpa',y='package',data=data)
plt.plot(data['cgpa'],le.predict(x),color='red')
# plt.savefig('plot.png')
plt.show()
```
