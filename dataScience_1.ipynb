for i in range(...) → “Go through each number and do something”

[i for i in range(...)] → “Go through each number and save it in a list(it is called list comprehension)
It is equivalent to writing this longer code:
result = []
for i in range(0, 81, 10):
    result.append(i)

result


[mn for i in range(0,300)] gives [29, 29, 29, 29, ..., 29]   # 300 times and
[i for i in range(0,300)]

This becomes:

[0, 1, 2, 3, ..., 299]
the result is like:plt.plot(x, y)

Matplotlib plots points like:

(29, 0)
(29, 1)
(29, 2)
...
(29, 299)


```python
plt.figure(figsize=(6,5))
sns.histplot(x='Age',data=dataset,bins=[i for i in range(0,81,10)])
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],color='orange')
plt.show()
```

```python
dataset.isnull().sum()
```

```python
dataset['Age'].fillna(dataset['Age'].mean(),inplace=True)
dataset.isnull().sum()
```

**Median**


```python
dataset['Age'].median()
```

```python
md=np.median(dataset['Fare'])
md
```

```python
plt.figure(figsize=(6,5))
sns.histplot(x='Fare',data=dataset,bins=[i for i in range(0,100,10)])
plt.plot([md for i in range(0,500)],[i for i in range(0,500)],color='yellow')
plt.show()
```

**Mode (most frequent data)** most of time we check mode for categorical data and also check for numerical data also


```python
type(dataset['Fare'].mode())#default this is pandas series datatype that why if we add [0] in that we get exact 0th index element
```

```python
mo=dataset['Fare'].mode()[0]
mo
```

```python
dataset['Fare'].value_counts()#it is used for frequency of the values
```

```python
plt.figure(figsize=(6,5))
sns.histplot(x='Fare',data=dataset,bins=[i for i in range(0,100,10)])
plt.plot([mo for i in range(0,500)],[i for i in range(0,500)],color='red',label='mode')
plt.plot([mn for i in range(0,300)],[i for i in range(0,300)],color='orange',label='mean')
plt.plot([md for i in range(0,500)],[i for i in range(0,500)],color='yellow',label='median')
plt.legend()
plt.show()
```

2.MAD- meand absolute diviation


```python
secA=np.array([75,65,73,68,72,76])
secB=np.array([90,47,43,96,93,51])
arr=np.array([1,2,3,4,5,6])
mean=np.mean(secA)
mean
# np.mean(secB)
```

```python
plt.figure(figsize=(8,4))
plt.scatter(secA, arr, color='orange', label='class A')
plt.scatter(secB, arr, label='class B')
plt.plot([70,70,70,70,70,70],[1,2,3,4,5,6])
plt.legend()
plt.show()
```

```python
mad_a=np.sum(np.abs(secA-mean))/len(secA)
mad_a#mad is less in sec A
```

```python
mad_b=np.sum(np.abs(secB-mean))/len(secB)
mad_b#mad is greater than sec A
```
