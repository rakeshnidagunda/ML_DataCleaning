```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
data={
    "Name":['cat','dog','cow','parrot']
}
data=pd.DataFrame(data)
data
```

```python
le=LabelEncoder()
data['label_encoded']=le.fit_transform(data['Name'])
data
```
