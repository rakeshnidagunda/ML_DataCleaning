```python
import pandas as pd
```

```python
data={
    "size":['s','m','xl','xxl','l','m','s','xxl','m']
}
data=pd.DataFrame(data)
data
```

```python
ord_df=[['s','m','l','xl','xxl']]
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder(categories=ord_df)
oe.fit(data[['size']])
```

```python
data['size_new']=oe.transform(data[['size']])
data
```

ordinal encoding using map function

```python
ord_mp={
    's':0,'m':1,'l':2,'xl':3,'xxl':4
}#if we want any random number, we able to give
```

```python
data['size_mp']=data['size'].map(ord_mp)
data
```
