Handling Duplicate data


```python
import pandas as pd
```

```python
data={
    "name":['a','b','c','d','a','c'],
    'eng':[8,7,5,8,8,4],
    'kannada':[2,3,4,5,2,6]
}
df=pd.DataFrame(data)
df
```

```python
df.duplicated()
```

```python
# df.drop_duplicates(subset=['name'], inplace=True)#using this we delete data row by single name also
df.drop_duplicates(inplace=True,keep='first')
df
```
