Feature scaling:
1. standardization
2. normalization

**Standardization:** its a very effective technique which resclae a feature value so that it has distribution with 0 mean value and variance equals to 1.
Xnew=(xi-xmean)/std


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import files
files.upload_file('loan.csv')
```

```python
dataset=pd.read_csv('loan.csv')
dataset.head(3)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset['ApplicantIncome'])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[['ApplicantIncome']])
```

```python
dataset['ApplicantIncome_ss']=ss.transform(dataset[['ApplicantIncome']])
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title('Before Scaling')
sns.distplot(dataset['ApplicantIncome'])
plt.subplot(1,2,2)
plt.title('After scaling')
sns.distplot(dataset['ApplicantIncome_ss'])
plt.show()
```

```python
dataset.describe()
```

**Normalization:**
its a scaling technique in which value are shifted and rescaled so that they end up ranging between 0 and 1. it is also known as min max scaling.
Xnew=(xi-min(X))/(max(X)-min(X)

```python
sns.distplot(dataset['CoapplicantIncome'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
mms=MinMaxScaler()
mms.fit(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_mms']=mms.transform(dataset[['CoapplicantIncome']])
```

```python
dataset['CoapplicantIncome_mms']
```

```python
plt.figure(figsize=(10,3))
plt.subplot(1,2,1)
plt.title('before scaling')
sns.distplot(dataset['CoapplicantIncome'])
plt.subplot(1,2,2,)
plt.title('after scaling')
sns.distplot(dataset['CoapplicantIncome_mms'])
plt.show()
```
