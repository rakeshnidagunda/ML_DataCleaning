KMeans clustering algorithm


```python
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
data={
    'customer':['riya','aman','fizan','neha','imran','sneha'],
    'age':[25,30,45,35,22,23],
    'spending':[100,200,300,110,290,300]
}
df=pd.DataFrame(data)
```

```python
x=df[['age','spending']]
model=KMeans(n_clusters=2,random_state=42,n_init=10)
df['group']=model.fit_predict(x)
print(df)
```

```python
plt.figure(figsize=(6,5))
for group in df['group'].unique():
  data_group=df[df['group']==group]#boolean masking
  plt.scatter(data_group['age'],data_group['spending'],label=f'group{group}')
plt.xlabel('age')
plt.ylabel('spending')
plt.title('customer segmentation')
plt.legend()
plt.show()
```

PCA-> principle componert analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
data={
    'age':[25,30,45,35,21],
    'income':[3000,4000,2000,1000,4000],
    'spending':[90,10,30,45,29],
    'saving':[1000,2000,3000,2000,2500]
}
df=pd.DataFrame(data)
print(df)
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df)
print(scaled_data)
```

```python
pca=PCA(n_components=2)
pca_result=pca.fit_transform(scaled_data)
print(pca_result)
```

```python
pca_df=pd.DataFrame(data=pca_result,columns=['PC1','PC2'])
print(pca_df)
```

```python
expalined_variance=pca.explained_variance_ratio_
print(expalined_variance)#variance captured by each pca
print(np.round(expalined_variance*100,2))
```

```python
plt.figure(figsize=(6,5))
plt.scatter(pca_df['PC1'],pca_df['PC2'],color='black',label="pca")
plt.xlabel('pca1')
plt.ylabel('pca2')
plt.title('PCA analysis')
plt.legend()
plt.grid(True)
plt.show()
```
