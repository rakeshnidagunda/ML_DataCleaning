A feature selection is an atribute that has an impact on a problem or is useful for the problem, and choosing the important features for the model is known as feature selection.

```python
import pandas as pd
from mlxtend.feature_selection import SequentialFeatureSelector as sfs
```

```python
from google.colab import files
files.upload_file('diabetes.csv')
```

```python
data=pd.read_csv('diabetes.csv')
data.head(3)
```

```python
x=data.iloc[:,:-1]
y=data['Outcome']
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
```

```python
fs=sfs(lr,k_features=6,forward=True)
fs.fit(x,y)
```

```python
fs.feature_names
```

```python
fs.k_score_
```
