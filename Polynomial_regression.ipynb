```python
data=pd.read_csv('polynomial.csv')
data.head(3)
```

```python
plt.scatter(data['Temperature (°C)'],data['Ice Cream Sales (units)'])
plt.show()
```

polynomial regression is a regression algorithm that models the relationship between a dependent(y) and independent variable (x) as nth degreee polynomial.
y=b0+b1x1**1+b2x2**2+.....bnxn**n

```python
data.corr()
```

```python
x=data[['Ice Cream Sales (units)']]
y=data['Temperature (°C)']
```

```python
from sklearn.preprocessing import PolynomialFeatures
pf=PolynomialFeatures(degree=2)
pf.fit(x)
pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)
```

```python
pred=lr.predict(x)
```

```python
plt.scatter(data['Temperature (°C)'],data['Ice Cream Sales (units)'])
plt.plot(data['Temperature (°C)'],pred,color='red')
plt.show()
```
